```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from datetime import date
```

# house_subs_blk1 house 1

```python
h1sb1 = pd.read_csv('house1_subs_blk1.csv')

# Convertissons la colonne unix_ts en datetime et passons cette colonne en index

h1sb1['unix_ts'] = pd.to_datetime(h1sb1['unix_ts'], unit='s')
h1sb1 = h1sb1.set_index('unix_ts')

h1sb1.head(40)
```

## Regardons pour le sub1

```python
df = h1sb1.loc[h1sb1['sub']==1]
df
```

# Graphiques Sub1

```python
df = df.iloc[:,1:].copy()

for i in range(df.shape[1]):
    fig = df[df.columns[i]].plot(figsize=(15,5), title=df.columns[i])
    fig.axes.title.set_size(20)
    plt.show()
```
